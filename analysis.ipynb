```python
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv("./attacks.csv", encoding="ISO-8859-1")
df.columns
df
```

```python
df.dropna( axis = 0, how = "all")
df.head()
```

```python
columnas = ["Date", "Country", "Area", "Location", "Activity", "Name", "Injury", "Fatal (Y/N)", "Species "]
corte1 = df[columnas]
corte1.head(45)
```

```python
diccio_columnas= {"Date" : "Date", "Country" : "Country", "Area" :"Area", "Location" : "Location", "Activity" :"Activity", "Name" : "Name", "Injury" : "Injury", "Fatal (Y/N)" : "Lethal", "Species " : "Species"}
```

```python
corte1 = corte1.rename(columns = diccio_columnas)
corte1
```

```python
corte1.iloc[0]
```

```python
corte1.dropna( axis = 0, inplace = True)
corte1
```

```python
filtrado = corte1.loc[(corte1["Country"] == "USA") & (corte1["Lethal"] == "Y") & (corte1["Activity"] == "Surfing") & (corte1["Species"] == "White shark" )]
filtrado.head(22)
```
